```python
import pandas as pd

# 1. Load the marketing dataset
df = pd.read_csv("85334965-5736-457a-b8d4-a077e6872f84 (1).csv")

# 2. Check how many missing values exist before cleaning
print("Missing values before cleaning:")
print(df.isnull().sum())

# 3. Clean the dataset by dropping rows with missing values
df = df.dropna()

# 4. Verify that missing values are completely gone
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Preview the first few rows of your clean data
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Calculate the mathematical correlation between columns
correlation_matrix = df.corr()
print("--- Correlation with Sales ---")
print(correlation_matrix["Sales"])

# 2. Visualize the relationships using a pairplot
sns.pairplot(df, x_vars=["TV", "Radio", "Social_Media"], y_vars="Sales", height=4, aspect=1, kind="reg")
plt.show()
```

```python
import statsmodels.api as sm

# 1. Define your Independent Variable (X) and Dependent Variable (y)
X = df["TV"]
y = df["Sales"]

# 2. Add a constant (intercept) to the model - statsmodels requires this explicitly
X = sm.add_constant(X)

# 3. Fit the Ordinary Least Squares (OLS) regression model
model = sm.OLS(y, X).fit()

# 4. Print the comprehensive results summary
print(model.summary())
```

```python
import numpy as np

# 1. Extract the fitted values and residuals from your model
fitted_values = model.fittedvalues
residuals = model.resid

# 2. Set up a beautiful plotting area
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot A: Residuals vs. Fitted Values (Tests Linearity and Homoscedasticity)
sns.scatterplot(x=fitted_values, y=residuals, ax=axes[0], alpha=0.5, color="purple")
axes[0].axhline(y=0, color='red', linestyle='--')
axes[0].set_title('Residuals vs. Fitted Values')
axes[0].set_xlabel('Fitted (Predicted) Sales')
axes[0].set_ylabel('Residuals (Errors)')

# Plot B: Histogram of Residuals (Tests Normality of Errors)
sns.histplot(residuals, kde=True, ax=axes[1], color="teal")
axes[1].set_title('Distribution of Residuals')
axes[1].set_xlabel('Residual Error Value')

plt.tight_layout()
plt.show()
```
